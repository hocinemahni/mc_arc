```python
import matplotlib.pyplot as plt

# 1. PARAMÈTRES GLOBAUX : Tout en gras et police agrandie par défaut
plt.rcParams.update({
    'font.weight': 'bold',
    'font.size': 14,          # Taille de base augmentée
    'axes.labelweight': 'bold',
    'axes.titleweight': 'bold'
})

# J'ai légèrement agrandi la figure (14, 5) pour que le gros texte ne se chevauche pas
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Données harmonisées
data = {
    'Yombo': {'labels': ['8', '10', '> 10'], 'values': [2, 29, 69]},
    'incompact3d': {'labels': ['2', '> 2'], 'values': [15, 85]},
    'NAMD': {'labels': ['3', '> 3'], 'values': [15, 85]}
}

bar_color = '#20639B'

for i, (title, d) in enumerate(data.items()):
    ax = axes[i]
    bars = ax.bar(d['labels'], d['values'], color=bar_color, edgecolor='white', width=0.6)

    # Titres des sous-graphiques (encore plus gros)
    ax.set_title(title, fontweight='bold', color='#1A4368', fontsize=18)

    # Axe X
    ax.set_xlabel('Number of accesses', fontweight='bold', fontsize=16)

    # Axe Y
    if i == 0:
        ax.set_ylabel('Files (%)', fontweight='bold', fontsize=16)

    # Mettre en gras les valeurs (ticks) sur les axes X et Y
    ax.tick_params(axis='both', labelsize=14)
    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_fontweight('bold')

    ax.set_ylim(0, 100)
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    ax.set_axisbelow(True)

    # Pourcentages sur les barres
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval}%',
                ha='center', va='bottom', fontweight='bold', color='#333333', fontsize=15)

# Légende principale
plt.suptitle('Figure 1: Distribution of files by number of accesses in three production traces',
             y=0.02, fontsize=18, fontweight='bold')

plt.tight_layout()
# On laisse un peu plus de place en bas pour le titre et les gros labels
plt.subplots_adjust(bottom=0.25)
plt.show()
```
